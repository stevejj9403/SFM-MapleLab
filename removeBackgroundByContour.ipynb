```python
import cv2
import numpy as np
import os

outputDirectory = "balloonNoBG"

#CHANGE THE directory_path AND THE os.makedirs()

directory_path = "C:\\Users\\steve\\Vanderbilt\\Research\\balloonImgs"


try:     
# creating a folder named data
    if not os.path.exists(outputDirectory):
        os.makedirs(outputDirectory)

        
# if not created then raise error
except OSError:
    print ('Error: Creating directory of data')


for filename in os.listdir(directory_path):
    print('removing background from ' + filename)
    if filename.endswith(".jpg"): # or filename.endswith(".png"): 
        image_path = os.path.join(directory_path, filename)
        img = cv2.imread(image_path)

        # Convert the image to grayscale
        gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Apply a threshold to the image
        _, threshold = cv2.threshold(gray_image, 120, 255, cv2.THRESH_BINARY_INV)

        # Find the contours of the binary image
        contours, _ = cv2.findContours(threshold, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

        # Find the largest contour
        largest_contour = max(contours, key=cv2.contourArea)

        # Create a mask of the object
        mask = np.zeros(img.shape[:2], dtype=np.uint8)
        cv2.drawContours(mask, [largest_contour], -1, (255, 255, 255), -1)

        # Apply the mask to the original image
        masked_image = cv2.bitwise_and(img, img, mask=mask)

        # Save the result
        img_name = './' + outputDirectory+'/' + filename
        cv2.imwrite(img_name, masked_image)
        
```
